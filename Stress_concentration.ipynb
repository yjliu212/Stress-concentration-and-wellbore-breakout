```python
import numpy as np
import matplotlib.pyplot as plt

angle = np.arange(0, 180, 1) # azimuth from the Shmax direction
Shmax = 13000
Shmin = 7000
pp = 4000
pmud = 4000

R = 1 # Normalized wellbore radius
R = R**2 # R square

theta = np.pi * angle / 180  # Convert angle to radians using numpy
EffHoopS = 0.5 * (Shmax + Shmin - 2 * pp) * (1 + 1 / R) - 0.5 * (Shmax - Shmin) * (1 + 3 / R ** 2) * np.cos(
    2 * theta) - (pmud - pp) / R

plt.figure()
plt.plot(angle, EffHoopS)  # Plot against angle, not R
plt.plot(np.array([-5,180]),np.array([20000,20000]),'--',label='Strong Rock, C0=20000 psi')
plt.plot(np.array([-5,180]),np.array([15000,15000]),'--',label='Strong Rock, C0=15000 psi')
plt.legend()
plt.xlabel('Angle (degrees)')  # Correct x-axis label
plt.ylabel('Effective Hoop Stress (psi) at wellbore wall')
plt.show()

PMUD=np.arange(4000,10000,1000)

EHS=np.zeros((len(theta),len(PMUD)))
for i in range(len(PMUD)):
    EHS[:,i]=0.5 * (Shmax + Shmin - 2 * pp) * (1 + 1 / R) - 0.5 * (Shmax - Shmin) * (1 + 3 / R ** 2) * np.cos(
    2 * theta) - (PMUD[i] - pp) / R

plt.figure()
plt.plot(angle, EHS,label=['MW = 4000 psi','MW = 5000 psi','MW = 6000 psi','MW = 7000 psi','MW = 8000 psi','MW = 9000 psi'])  # Plot against angle, not R
plt.plot(np.array([-5,180]),np.array([15000,15000]),'g--')
plt.legend()
plt.xlabel('Angle (degrees)')  # Correct x-axis label
plt.ylabel('Effective Hoop Stress (psi) at wellbore wall')
plt.show()

# print(PMUD)
```
